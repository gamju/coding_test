```python
from collections import deque

#2178
def find_nearest(cur_coord, map_info, map_metric):
    y,x = cur_coord[0], cur_coord[1]
    next_coord_list = []
    cur_metric = map_metric[y][x]
    map_info[y][x] = 0
    direction_key = {"left" : (0,-1), "right" : (0,1),"top" : (-1,0),"bottom": (1,0)}
    for direction in direction_key.keys():
        try:
            next_y, next_x = y+direction_key[direction][0], x+direction_key[direction][1]
            if next_y < 0 or next_x < 0:
                continue
#             print(next_y, next_x, cur_metric)
            if map_info[next_y][next_x] == 1:
#                 print(next_y,next_x)
                metric = cur_metric + 1
                if metric < map_metric[next_y][next_x]:
                    map_metric[next_y][next_x] = metric
#                     map_info[next_y][next_x] = 1
                    next_coord_list.append((next_y, next_x))                
        except:
            continue
#     print(next_coord_list)
    return next_coord_list, map_info, map_metric

def miro():
    N,M = list(map(int, input().split(" ")))
#     N,M = 2,25
#     row = ["101111", "101010", "101011", "111011"]
#     row = ["110110", "110110", "111111", "111101"]
#     row = ["1011101110111011101110111", "1110111011101110111011101"]
    maxmimum_value = M*N
    map_metric = []
    map_info = []
    coord_queue = deque()
    coord_queue.append((0,0))
    #map initialization
    for i in range(N):
        row_info = list(map(int, list(input())))
#         row_info = list(map(int, list(row[i])))
        map_info.append(row_info)
        map_metric.append([maxmimum_value for _ in range(M)])
    map_metric[0][0] = 1
    while(True):
#         print(N, M)
        cur_coord = coord_queue.popleft()
        if N-1 == cur_coord[0] and M-1 == cur_coord[1]:
            print(map_metric[N - 1][M - 1])
            break
        find_coord, map_info, map_metric = find_nearest(cur_coord, map_info, map_metric)
#         print(find_coord)
        coord_queue.extend(find_coord)
    
    return N

# list_map = list(map(int,list('222')))
# print(list_map)
# coord_queue = deque()
# coord_queue.append((1,1))
# coord_queue.extend([(1,1),(1,1),(1,1)])
# print(coord_queue)

miro()
```
